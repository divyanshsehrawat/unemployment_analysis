```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment in India.csv")
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = ['Region', 'Date', 'Frequency', 'Unemployment Rate (%)', 'Employed', 'Labour Participation Rate (%)', 'State']
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Unemployment Rate (%)', y='State', data=df.sort_values('Unemployment Rate (%)', ascending=False))
plt.title("Unemployment Rate by State")
plt.show()
```

```python
# Heatmap by region
region_data = df.groupby('Region').mean(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(region_data[['Unemployment Rate (%)']], annot=True, cmap='Reds')
plt.title("Average Unemployment Rate by Region")
plt.show()
```
